Normalization of Array's

```python
#Normalization of Array's

from sklearn import preprocessing
import pandas as pd
import numpy as np
dataset = pd.read_csv('USA_cars_datasets.csv')

year = np.array(dataset['year'])
price=np.array(dataset['price'])
mileage = np.array(dataset['mileage'])
lot = np.array(dataset['lot'])

normalized_array = preprocessing.normalize([year,price,mileage,lot])
print(normalized_array)

dataset['year']= normalized_array[0]
dataset['price']= normalized_array[1]
dataset['mileage']= normalized_array[2]
dataset['lot']= normalized_array[3]

dataset
```

Encode categorical features in your data.


```python
df = pd.read_csv('USA_cars_datasets.csv')
df.head()
```

```python
df_categorical = df[['brand', 'model', 'title_status', 'vin', 'state','country','condition']]
df_categorical.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded = encoder.fit_transform(df_categorical)

feature_names = encoder.get_feature_names_out(input_features=['brand', 'model', 'title_status', 'vin', 'state','country','condition'])
education_encoded = pd.DataFrame(encoded, columns=feature_names)
education_encoded
print(
    f"The dataset is composed of {df_categorical.shape[1]} features")
df_categorical.head()
data_encoded = encoder.fit_transform(df_categorical)


print(
    f"The encoded dataset contains {data_encoded.shape[1]} features")

columns_encoded = encoder.get_feature_names_out(df_categorical.columns)
pd.DataFrame(data_encoded, columns=columns_encoded).head()
```

Transform skewed features into normally distributed features using transformation.

```python
from sklearn.preprocessing import QuantileTransformer

quantile = QuantileTransformer(output_distribution='normal')
year = np.array(dataset['year']).reshape(-1,1)
z= quantile.fit_transform(year)
sns.histplot(z, bins = 50, kde= True)
```

```python
from sklearn.preprocessing import QuantileTransformer

quantile = QuantileTransformer(output_distribution='normal')
price = np.array(dataset['price']).reshape(-1,1)
z= quantile.fit_transform(price)
sns.histplot(z, bins = 50, kde= True)
```
