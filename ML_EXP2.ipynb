Explore size, shape, data types of each column in the dataset.

```python
import pandas as pd
salary = pd.read_csv("USA_cars_datasets.csv")

for col,dtype in salary.dtypes.items():
  print(col, dtype)
print(salary.size)
print(salary.shape)
```

Identify null values, missing values in dataset.

```python
salary.isnull()
```

```python
print(salary.isna())
```

Replace missing values by appropriate values.

```python
salary.fillna(method="pad")
```

Remove null values from the dataset.

```python
salary.dropna
```
